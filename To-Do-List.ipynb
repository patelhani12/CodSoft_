# task:3 To-Do List

```python
import tkinter as tk
from tkinter import messagebox

class TodoListApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Todo List App")

        # Create a list to store tasks
        self.tasks = []

        # Create GUI components
        self.task_entry = tk.Entry(root, width=40)
        self.task_listbox = tk.Listbox(root, selectmode=tk.SINGLE, width=40)
        self.task_listbox.bind('<<ListboxSelect>>', self.on_task_select)

        # Buttons
        self.add_button = tk.Button(root, text="Add Task", command=self.add_task)
        self.update_button = tk.Button(root, text="Update Task", command=self.update_task)
        self.delete_button = tk.Button(root, text="Delete Task", command=self.delete_task)

        # Place GUI components using grid layout
        self.task_entry.grid(row=0, column=0, padx=10, pady=10, columnspan=3)
        self.task_listbox.grid(row=1, column=0, padx=10, pady=10, columnspan=3)
        self.add_button.grid(row=2, column=0, padx=10, pady=10)
        self.update_button.grid(row=2, column=1, padx=10, pady=10)
        self.delete_button.grid(row=2, column=2, padx=10, pady=10)

    def add_task(self):
        task = self.task_entry.get()
        if task:
            self.tasks.append(task)
            self.update_task_listbox()
            self.clear_entry()

    def update_task(self):
        selected_index = self.task_listbox.curselection()
        if selected_index:
            task = self.task_entry.get()
            if task:
                self.tasks[selected_index[0]] = task
                self.update_task_listbox()
                self.clear_entry()

    def delete_task(self):
        selected_index = self.task_listbox.curselection()
        if selected_index:
            del self.tasks[selected_index[0]]
            self.update_task_listbox()
            self.clear_entry()

    def on_task_select(self, event):
        selected_index = self.task_listbox.curselection()
        if selected_index:
            selected_task = self.tasks[selected_index[0]]
            self.task_entry.delete(0, tk.END)
            self.task_entry.insert(0, selected_task)

    def update_task_listbox(self):
        self.task_listbox.delete(0, tk.END)
        for task in self.tasks:
            self.task_listbox.insert(tk.END, task)

    def clear_entry(self):
        self.task_entry.delete(0, tk.END)

if __name__ == "__main__":
    root = tk.Tk()
    app = TodoListApp(root)
    root.mainloop()
```
